```python
import pandas as pd
import json
import gzip

# File path
file_path = '/content/drive/My Drive/Bahar_research/meta_Movies_and_TV.json.gz'

chunk_size = 10000  # Number of lines per chunk
chunks = []

# Open and read the .gz file
with gzip.open(file_path, 'rt') as f:
    chunk = []
    for i, line in enumerate(f):
        chunk.append(json.loads(line))  # Parse each JSON object
        if (i + 1) % chunk_size == 0:
            chunks.append(pd.DataFrame(chunk))  # Convert chunk to DataFrame
            chunk = []  # Reset chunk

    # Process any remaining lines
    if chunk:
        chunks.append(pd.DataFrame(chunk))

# Combine all chunks into a single DataFrame
m_meta_df = pd.concat(chunks, ignore_index=True)
```

#### Keeping only similar review id rows

```python
# Read unique 'asin' values from the file
with open(r"/content/unique_reviewerID_book.txt", 'r') as f:
    reviewerID_set = set(f.read().splitlines())

# Assuming 'df' is the pre-loaded DataFrame
chunksize = 10**6  # Define the chunk size for processing
filtered_rows = []  # To store filtered data temporarily

# Loop through the DataFrame in chunks
for start in range(0, len(merged_df), chunksize):
    # Get a chunk of the DataFrame
    chunk = merged_df.iloc[start:start + chunksize]

    # Process the chunk using .loc to avoid SettingWithCopyWarning
    chunk.loc[:, 'reviewerID'] = chunk['reviewerID'].astype(str).str.strip()

    # Filter the chunk based on reviewerID
    filtered_chunk = chunk[chunk['reviewerID'].isin(reviewerID_set)]

    # Append the filtered chunk to the list
    filtered_rows.append(filtered_chunk)

# Combine all filtered chunks into a single DataFrame
filtered_df_movie = pd.concat(filtered_rows, ignore_index=True)

# Optionally, save the filtered DataFrame to a new file
# filtered_df.to_csv('filtered_data.csv', index=False)

print(f"Filtered DataFrame has {len(filtered_df_movie)} rows.")
```
